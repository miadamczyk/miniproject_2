# Wprowadzenie

Głębokie sieci neuronowe osiągają bardzo wysoką skuteczność w zadaniach computer vision, jednak większość tych modeli działa jako black-box, przez co trudno zrozumieć, dlaczego model podjął daną decyzję oraz ciężko zweryfikować, czy model wykorzystuje właściwe cechy danych.

Explainable Artificial Intelligence (XAI) zajmuje się problemem interpretacji decyzji modeli uczenia maszynowego.

# Problem modeli typu Black-Box

Modele deep learning uczą się bardzo złożonych reprezentacji danych, często w niezrozumiały sposób. Ich pipeline zwykle wygląda następująco:

input image → deep neural network → prediction

Reprezentacje danych generowane przez sieć są wysokowymiarowe i nie mają bezpośredniego znaczenia semantycznego, przez co ich interpretacja w naiwny sposób jest trudna lub niemożliwa.

Powoduje to problemy w zastosowaniach takich jak:

- autonomiczne pojazdy
- systemy bezpieczeństwa
- analiza danych wrażliwych
- medycyna


które często, w zastosowaniach komercyjnych, wymagają uzasadnienia działania modelu, dlatego umożlienie interpretowalności jest tak ważnym zagadnieniem.

# Główne podejścia do XAI w Computer Vision

Cztery główne podejścia do wyjaśnialności w wizji komputerowej wskazane w artykule:

### 1. Saliency Maps (mapy istotności)

Metody pokazujące, które piksele obrazu miały największy wpływ na predykcję.

### 2. Concept Bottleneck Models (modele z wąskim gardłem pojęciowym)

Modele wykorzystujące interpretowalną warstwę pojęć.

image -> predict concepts -> [concept, concept, ...] -> classifier -> prediction

### 3. Prototype-based methods (metody prototypowe)

Modele porównujące nowe dane do przykładów prototypowych z danych treningowych.

### 4. Hybrid methods (metody hybrydowe)

Metody łączące różne podejścia interpretowalności.

## Saliency Maps 

Celem tych metod jest pokazanie, które fragmenty obrazu były najważniejsze dla predykcji modelu. Saliency map przedstawiana jest zazwyczaj jako heatmapa nałożona na obraz.
Pipeline:

image -> model deep learning -> prediction + sailency map

Mapa pokazuje gdzie model "patrzył" podczas podejmowania decyzji. Intuicyjnie, jeśli zmiana jakiegoś piksela silnie wpływa na wynik modelu, to piksel ten jest ważny dla predykcji (tzn. "Jak bardzo predykcja modelu zmieni się, jeśli minimalnie zmienimy dany piksel?").

Pierwsze modele saliency maps trenowane były jedynie na etykietach klas a mapa istotności była obliczana po treningu, na podstawie gradientów.

Podejście to ma jednak kilka ograniczeń:

- mapy są często rozmyte
- mogą być wrażliwe na niewielki szum w obrazie
- czasem podświetlają regiony niezwiązane z obiektem

## Zalety i ograniczenia Saliency Maps

### Zalety

**• Łatwość zastosowania**  
Większość metod saliency, szczególnie metod gradientowych, można bezpośrednio zastosować
do już wytrenowanych modeli CNN. Nie wymagają one zmiany architektury modelu ani ponownego treningu.

**• Intuicyjna interpretacja**  
Wyniki przedstawiane są w postaci heatmap które można łatwo interpretować wizualnie.
Pozwala to szybko zidentyfikować obszary obrazu, na które model zwraca największą uwagę.

**• Liczne metody**  
Od podstawowych metod gradientowych powstało wiele rozszerzeń poprawiających
stabilność i interpretowalność, np.:

- CAM / Grad-CAM  
- SmoothGrad  
- Integrated Gradients  

**• Szerokie zastosowanie**  
Saliency maps znajdują zastosowanie w wielu dziedzinach.

---

### Ograniczenia

**• Niestabilność wyników**  
Mapy saliency mogą być wrażliwe na szum w danych wejściowych
lub niewielkie zmiany modelu. Jednocześnie, eksperymenty pokazują, że czasami podobne
mapy powstają nawet po losowej zmianie wag modelu.

**• Brak dobrych metryk weryfikacji działania modeli**  
Istnieją różne metryki oceny (np. AOPC, Deletion/Insertion, Pointing Game),
jednak żadna z nich nie jest uznawana za uniwersalny standard.

**• Ryzyko nadinterpretacji**  
Heatmapy są często rozmyte i nieprecyzyjne, co może prowadzić
do błędnej interpretacji zaznaczonych obszarów.

**• Brak poziomu semantycznego (concept level)**  
Saliency maps wskazują jedynie ważne piksele lub regiony obrazu,
ale nie wyjaśniają decyzji modelu w kategoriach pojęć semantycznych
(np. „kształt dzioba”, „kolor skrzydeł”).

## Concept Bottleneck Models

Kolejnym z podejść do zwiększenia interpretowalności modeli jest wykorzystanie Concept Bottleneck Models (CBM). Idea polega na wprowadzeniu warstwy pojęć semantycznych (concepts) pomiędzy danymi wejściowymi a predykcją.

## Intuicja działania CBM

CBM rozbijają proces predykcji na dwa etapy:

1. Predykcja konceptów - model analizuje obraz i przewiduje wartości zestawu pojęć opisujących dane.

    image -> predict concepts -> [ concept1, concept2, ... ]

2. Predykcja klasy - model wykorzystuje pojęcia do wyznaczenia końcowej klasy:

    [ concept1, concept2, ... ] -> classifier -> prediction = class_i

    Dzięki temu decyzja modelu może być łatwo interpretowana.

### Jak model rozumie koncepty ?

W CBM dane treningowe zawierają nie tylko etykietę klasy, ale również etykiety konceptów.

#### Przykład datasetu

| obraz | czerwone skrzydła | długi dziób | biały brzuch | gatunek |
|------|------------------|------------|-------------|--------|
| img1 | 1 | 0 | 1 | cardinal |
| img2 | 0 | 1 | 0 | heron |

Koncepty opisują cechy, które są zrozumiałe dla człowieka.

---

### Działanie modelu

Dwa etapy:

#### 1. Predykcja konceptów

Model analizuje obraz i przewiduje wartości konceptów, np. możemy uzyskać wynik [0.8, 0.1, 0.9], co oznacza:

- czerwone skrzydła ≈ 0.8  
- długi dziób ≈ 0.1  
- biały brzuch ≈ 0.9  

---

#### 2. Predykcja klasy

Drugi model, na podstawie przewidzianych konceptów klasyfikuje dane wejściowe. Najczęściej jest to prosty klasyfikator (np. model liniowy lub mała sieć).


W klasycznych CBM koncepty są zdefiniowane przez człowieka, oznaczone w danych treningowych, dlatego jednym z głównych ograniczeń CBM jest kosztowna anotacja konceptów w datasetach.


## Praktyczne zastosowania Concept Bottleneck Models

CBM znajdują zastosowanie szczególnie w medycynie, gdzie interpretowalność modeli jest kluczowa.

W przeciwieństwie do saliency maps, które wskazują jedynie ważne piksele obrazu,
CBM wykorzystują koncepty, co jest bliższe rozumowaniu człowieka.

### Przykłady zastosowań

**Diagnostyka dermatologiczna**  
Modele CBM mogą klasyfikować zmiany skórne, wykorzystując koncepty, np. czerwony kolor, nieregularne brzegi.

Najpierw model rozpoznaje zadane cechy, a następnie klasyfikuje zmianę jako łagodną lub złośliwą.

**Analiza zdjęć RTG klatki piersiowej**  
CBM pozwala analizować związek między zadanymi konceptami a diagnozą, np. powiększenie serca, zacienienia płuc, obecność płynu.

Możliwa jest także modyfikacja pojęć przez lekarza i sprawdzenie, jak wpyłnie to na zawchowanie (predykcję) modelu.

**Wyjaśnienia multimodalne**  
Niektóre systemy łączą CBM z modelami językowymi,
aby generować tekstowe wyjaśnienia decyzji modelu
na podstawie wykrytych pojęć.

## Zalety i ograniczenia Concept Bottleneck Models

### Zalety

**• Interpretowalność poprzez koncepty**  
CBM wyjaśniają decyzje modelu przy użyciu konceptów zrozumiałych dla człowieka, zamiast abstrakcyjnych cech sieci neuronowej. Dzięki temu proces podejmowania decyzji jest bardziej przejrzysty, niż w poprzedniej metodzie.

**• Możliwość interwencji i korekty błędów**  
Użytkownik może bezpośrednio modyfikować wartości konceptów i obserwować, jak wpływa to na predykcję modelu. Umożliwia to:
- analizę błędów modelu  
- testowanie hipotez  
- zwiększenie zaufania do modelu 

**• Potencjał do analiz przyczynowych**  
CBM explicite modelują łańcuch decyzyjny. Dzięki temu możliwe jest prowadzenie analiz przyczynowych i kontrfaktycznych, a nie tylko analiz opartych na korelacjach.

---

### Ograniczenia

**• Konieczność definiowania konceptów**  
Skuteczność CBM zależy od zestawu konceptów zdefiniowanych przez człowieka. Wymaga to:
- dużego nakładu pracy
- wiedzy eksperckiej
- subiektywnego wyboru istotnych cech

**• Ryzyko uczenia się skrótów (tzw. shortcut learning)**  
Model może nauczyć się wykorzystywać przypadkowe korelacje między konceptami a etykietą, co prowadzi do wyjaśnień, które wyglądają poprawnie, ale nie odzwierciedlają rzeczywistego procesu decyzyjnego.

**• Problemy ze skalowalnością**  
W zadaniach z dużą liczbą abstrakcyjnych lub trudnych do zdefiniowania konceptów
proces ich definiowania i anotacji staje się bardzo kosztowny i trudny do skalowania.

## Prototype-based Methods

Prototype-based methods to podejście do wyjaśnialności modeli, w którym decyzja modelu
jest tłumaczona poprzez porównanie nowego przykładu z reprezentatywnymi przykładami
(prototypami) nauczonymi z danych treningowych. 

Idea odpowiada sposobowi rozumowania człowieka, tzn. "Ten obiekt należy do klasy X, ponieważ wygląda podobnie do tego przykładu z klasy X". Model, zamiast analizować gradienty lub abstrakcyjne cechy sieci neuronowej, porównuje obraz z zestawem prototypów reprezentujących cechy poszczególnych klas.

Schemat działania:

image -> backbone -> porównanie z prototypami -> similarity scores -> prediction


Podczas predykcji model sprawdza:

- które fragmenty obrazu są najbardziej podobne do prototypów
- jak silne jest to podobieństwo

Na tej podstawie podejmowana jest decyzja klasyfikacyjna.

### Praktyczne zastosowania modeli prototypowych

Dzięki swojej przejrzystości i intuicyjnym wyjaśnieniom modele prototypowe
znajdują zastosowanie w wielu dziedzinach wymagających interpretowalności.

**Medycyna i analiza obrazów medycznych**  
Modele takie jak ProtoPNet mogą klasyfikować obrazy medyczne
(np. RTG, obrazy histopatologiczne) oraz wyjaśniać diagnozę poprzez
porównanie z prototypowymi przykładami, np.:
"ten guz jest złośliwy, ponieważ jego struktura komórkowa przypomina
prototyp komórek nowotworowych".

**Klasyfikacja biologiczna i środowiskowa**  
W zadaniach szczegółowej klasyfikacji gatunków modele pomagają
zrozumieć, które cechy morfologiczne (np. kształt dzioba,
kolor piór) są wykorzystywane do odróżniania podobnych gatunków.

**Samochody autonomiczne i robotyka**  
Modele prototypowe mogą wyjaśniać, dlaczego obiekt został
zidentyfikowany np. jako pieszy, wskazując podobieństwo do
prototypu sylwetki człowieka. Pomaga to w analizie i debugowaniu
systemów bezpieczeństwa.

**Kontrola jakości w produkcji**  
Modele mogą wykrywać wadliwe produkty na linii produkcyjnej,
wyjaśniając decyzję poprzez podobieństwo do prototypów defektów,
np. "produkt jest wadliwy, ponieważ posiada rysę podobną
do prototypu uszkodzenia".

### Zalety i ograniczenia modeli prototypowych

#### Zalety

**• Intuicyjność wyjaśnień**  
Wyjaśnienia oparte na konkretnych przykładach są łatwe do zrozumienia,
ponieważ przypominają ludzkie porównywanie nowych obiektów
do wcześniej znanych przykładów.

**• Wbudowana interpretowalność (ante-hoc)**  
Mechanizm wyjaśniania jest częścią architektury modelu,
dzięki czemu prezentowane wyjaśnienia rzeczywiście odzwierciedlają
proces decyzyjny modelu, a nie jedynie jego przybliżoną interpretację.

**• Odniesienie do rzeczywistych przykładów**  
Prototypy są rzeczywistymi fragmentami danych treningowych,
co pozwala użytkownikowi bezpośrednio zobaczyć,
do jakich przykładów model porównuje dane wejściowe.

---

#### Ograniczenia

**• Konieczność modyfikacji architektury modelu**  
Metoda nie może być łatwo zastosowana do już wytrenowanych modeli.
Architektura musi być zaprojektowana tak, aby zawierała warstwę prototypów.

**• Trudność w uczeniu znaczących prototypów**  
Podczas treningu prototypy mogą nauczyć się cech przypadkowych
lub zaszumionych, które nie reprezentują rzeczywistych
i jednoznacznych cech danych.

**• Wysoki koszt obliczeniowy**  
Podczas predykcji model musi porównać dane wejściowe
z wieloma prototypami, co zwiększa koszt obliczeń,
szczególnie gdy liczba prototypów jest duża.

## Podsumowanie przedstawionych metod

W artykule xAI-CV: An Overview of Explainable Artificial Intelligence in Computer Vision
wyróżniono trzy główne podejścia do wyjaśniania decyzji modeli wizji komputerowej:

1. **Saliency Maps** – pokazują gdzie model "patrzy" w obrazie.
2. **Concept Bottleneck Models (CBM)** – wyjaśniają, jakie koncepty zdecydowały o predykcji.
3. **Prototype-based methods** – wyjaśniają do jakiego przykładu z danych podobny jest analizowany obraz. 


## Porównanie metod XAI

| Cecha | Saliency Maps | Concept Bottleneck Models | Prototype-based Methods |
|------|---------------|---------------------------|-------------------------|
| **Wyjaśnienie** | Piksele / regiony obrazu | Koncepty semantyczne | Przykłady z danych |
| **Intuicja** | Gdzie model patrzy? | Jakie cechy model wykorzystuje? | Do jakiego przykładu to jest podobne? |
| **Typ metody** | Post-hoc | Ante-hoc | Ante-hoc |
| **Wymagana zmiana architektury modelu** | Nie | Tak | Tak |
| **Forma wyjaśnienia** | Heatmapa ważnych pikseli | Wartości konceptów | Prototypy |
| **Interpretowalność dla człowieka** | Średnia | Wysoka | Wysoka |
| **Możliwość interwencji użytkownika** | Brak | Tak (poprzez modyfikację konceptów) | Ograniczona |


Każde z podejść odpowiada na inne pytanie dotyczące działania modelu,
dlatego w praktyce często łączy się je w podejścia hybrydowe, aby uzyskać
bardziej kompletne wyjaśnienia działania systemów AI.

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)
```

```python
transform = transforms.Compose([
    transforms.Resize((224,224)),
    transforms.ToTensor()
])

train_dataset = torchvision.datasets.CIFAR10(
    root="./data",
    train=True,
    download=True,
    transform=transform
)

test_dataset = torchvision.datasets.CIFAR10(
    root="./data",
    train=False,
    download=True,
    transform=transform
)

train_loader = torch.utils.data.DataLoader(
    train_dataset,
    batch_size=64,
    shuffle=True
)

test_loader = torch.utils.data.DataLoader(
    test_dataset,
    batch_size=64,
    shuffle=False
)
```

```python
import torch
import torchvision
import torch.nn as nn

torch.cuda.empty_cache()

model = torchvision.models.resnet18(pretrained=True)

# zamrożony backbone, tylko classifier
for param in model.parameters():
    param.requires_grad = False
for param in model.fc.parameters():
    param.requires_grad = True
# aby użyć Grad-CAM:
for param in model.layer4.parameters():  # ostatni blok ResNet
    param.requires_grad = True

# nowa warstwa klasyfikacyjna
model.fc = nn.Linear(model.fc.in_features, 10)

model = model.to(device)
```

```python
def evaluate(model, loader):

    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():

        for images, labels in loader:

            images = images.to(device)
            labels = labels.to(device)

            outputs = model(images)

            preds = torch.argmax(outputs, dim=1)

            correct += (preds == labels).sum().item()
            total += labels.size(0)

    return correct / total
```
